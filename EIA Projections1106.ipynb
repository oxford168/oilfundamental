```python
from cornerstone import *
```

```python
start='2017-01'
end='2019-12'
rcParams['figure.figsize'] = 16, 16
#注："n"月底出"n-2"月均数据，如2018年9月28日更新至2018年7月数据，2018年八月及以后为预测值
```

```python
#Crude Oil
USCRSTKP=eia_m(apicode['CRSTKP'],'USCRSTKP')[start:end]
USCRPRODP=eia_m(apicode['CRPRODP'],'USCRPRODP')[start:end]
USCRNIMPP=eia_m(apicode['CRNIMPP'],'USCRNIMPP')[start:end]
USCRUTIP=eia_m(apicode['CRUTIP'],'USCRUTIP')[start:end]

fig, axes = plt.subplots(nrows=4, ncols=1)
plot_projection_m(USCRSTKP,axes[0],"mb")
plot_projection_m(USCRPRODP,axes[1],"mb/d")
plot_projection_m(USCRNIMPP,axes[2],"mb/d")
plot_projection_m(USCRUTIP,axes[3],"%")
axes[0].set_title('U.S. Commercial Crude Oil Projection, Monthly')
```

```python
#Distillate Fuel Oil
USGOSTKP=eia_m(apicode['GOSTKP'],'USGOSTKP')[start:end]
USGOPRODP=eia_m(apicode['GOPRODP'],'USGOPRODP')[start:end]
USGONIMPP=eia_m(apicode['GONIMPP'],'USGONIMPP')[start:end]
USGODEMP=eia_m(apicode['GODEMP'],'USGODEMP')[start:end]

fig, axes = plt.subplots(nrows=4, ncols=1)
plot_projection_m(USGOSTKP,axes[0],"mb")
plot_projection_m(USGOPRODP,axes[1],"mb/d")
plot_projection_m(USGONIMPP,axes[2],"mb/d")
plot_projection_m(USGODEMP,axes[3],"mb/d")
axes[0].set_title('U.S. Distillate Fuel Oil Projection, Monthly')
```
