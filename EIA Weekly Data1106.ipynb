```python
from cornerstone import *
```

```python
start='2014-01-03'
end='2018-11-05'
```

```python
#Crude Oil
USCRSTKW=eia_w(apicode['CRSTKW'],'USCRSTKW')[start:end]
USCRNIMPW=eia_w(apicode['CRNIMPW'],'USCRNIMPW')[start:end]
USCRUTIW=eia_w(apicode['CRUTIW'],'USCRUTIW')[start:end]

rcParams['figure.figsize'] = 18, 15
fig, axes = plt.subplots(nrows=3, ncols=1)
plot_seasonal(USCRSTKW,axes[0],"kb")
plot_seasonal(USCRNIMPW,axes[1],"kb/d")
plot_seasonal(USCRUTIW,axes[2],"%")
axes[0].set_title('U.S. Commercial Crude Oil, Weekly')
axes[0].legend(np.unique(USCRSTKW.index.year.values),loc='upper center',bbox_to_anchor=(1.05, 0.5),fancybox=True, shadow=True)
```

```python
#Distillate Fuel Oil
USGOSTKW=eia_w(apicode['GOSTKW'],'USGOSTKW')[start:end]
USGOPRODW=eia_w(apicode['GOPRODW'],'USGOPRODW')[start:end]
USGODEMW=eia_w(apicode['GODEMW'],'USGODEMW')[start:end]
USGOEXPW=eia_w(apicode['GOEXPW'],'USGOEXPW')[start:end]

rcParams['figure.figsize'] = 18, 18
fig, axes = plt.subplots(nrows=4, ncols=1)
plot_seasonal(USGOSTKW,axes[0],"kb")
plot_seasonal(USGOPRODW,axes[1],"kb/d")
plot_seasonal(USGODEMW,axes[2],"kb/d")
plot_seasonal(USGOEXPW,axes[3],"kb/d")
axes[0].set_title('U.S. Distillate Fuel Oil, Weekly')
axes[0].legend(np.unique(USGOSTKW.index.year.values),loc='upper center',bbox_to_anchor=(1.05, 0.5),fancybox=True, shadow=True)
```
