```python
from cornerstone import *
```

```python
start='2014-01-03'
end='2018-10-30'
```

```python
#Crude Oil
crude_stk=eia_w('PET.WCESTUS1.W','Commercial Crude Oil Stock')[start:end]
crude_imp=eia_w('PET.WCRNTUS2.W','Crude Oil Net Import')[start:end]
crude_uti=eia_w('PET.WPULEUS3.W','Utilization of Refinery Operable Capacity')[start:end]

rcParams['figure.figsize'] = 18, 15
fig, axes = plt.subplots(nrows=3, ncols=1)
plot_seasonal(crude_stk,axes[0],"kb")
plot_seasonal(crude_imp,axes[1],"kb/d")
plot_seasonal(crude_uti,axes[2],"%")
axes[0].set_title('U.S. Commercial Crude Oil, Weekly')
axes[0].legend(np.unique(crude_stk.index.year.values),loc='upper center',bbox_to_anchor=(1.05, 0.5),fancybox=True, shadow=True)
```

```python
#Distillate Fuel Oil
diesel_stk=eia_w('PET.WDISTUS1.W','Distillate Fuel Oil Stock')[start:end]
diesel_prod=eia_w('PET.WDIRPUS2.W','Distillate Fuel Oil Production')[start:end]
diesel_cons=eia_w('PET.WDIUPUS2.W','Distillate Fuel Oil Consumption')[start:end]
diesel_exp=eia_w('PET.WDIEXUS2.W','Distillate Fuel Oil Export')[start:end]

rcParams['figure.figsize'] = 18, 18
fig, axes = plt.subplots(nrows=4, ncols=1)
plot_seasonal(diesel_stk,axes[0],"kb")
plot_seasonal(diesel_prod,axes[1],"kb/d")
plot_seasonal(diesel_cons,axes[2],"kb/d")
plot_seasonal(diesel_exp,axes[3],"kb/d")
axes[0].set_title('U.S. Distillate Fuel Oil, Weekly')
axes[0].legend(np.unique(diesel_stk.index.year.values),loc='upper center',bbox_to_anchor=(1.05, 0.5),fancybox=True, shadow=True)
```

```python
#Gasoline
gasoline_stk=eia_w('PET.WGTSTUS1.W','Total Motor Gasoline Stock')[start:end]
gasoline_prod=eia_w('PET.WGFRPUS2.W','Finished Motor Gasoline Production')[start:end]
gasoline_cons=eia_w('PET.WGFUPUS2.W','Finished Motor Gasoline Consumption')[start:end]
gasoline_exp=eia_w('PET.W_EPM0F_EEX_NUS-Z00_MBBLD.W','Finished Motor Gasoline Export')[start:end]

rcParams['figure.figsize'] = 18, 18
fig, axes = plt.subplots(nrows=4, ncols=1)
plot_seasonal(gasoline_stk,axes[0],"kb")
plot_seasonal(gasoline_prod,axes[1],"kb/d")
plot_seasonal(gasoline_cons,axes[2],"kb/d")
plot_seasonal(gasoline_exp,axes[3],"kb/d")
axes[0].set_title('U.S. Motor Gasoline, Weekly')
axes[0].legend(np.unique(gasoline_stk.index.year.values),loc='upper center',bbox_to_anchor=(1.05, 0.5),fancybox=True, shadow=True)
```
