```python
from cornerstone import *
```

```python
start='2014-01'
end='2019-12'
#注："n"月底出"n-2"月均数据，如2018年9月28日更新至2018年7月数据，2018年八月及以后为预测值
```

```python
#Crude Oil
crude_stk=eia_m('STEO.COSXPUS.M','commercial stock')[start:end]
crude_prod=eia_m('STEO.COPRPUS.M','production')[start:end]
crude_imp=eia_m('STEO.CONIPUS.M','net imports')[start:end]
crude_uti=eia_m('STEO.ORUTCUS.M','Refinery Distillation Utilization Factor')[start:end]

rcParams['figure.figsize'] = 16, 16
fig, axes = plt.subplots(nrows=4, ncols=1)
plot_seasonal_m(crude_stk,axes[0],"mb")
plot_seasonal_m(crude_prod,axes[1],"mb/d")
plot_seasonal_m(crude_imp,axes[2],"mb/d")
plot_seasonal_m(crude_uti,axes[3],"%")
axes[0].set_title('U.S. Commercial Crude Oil Projection, Monthly')
axes[0].legend(np.unique(crude_prod.index.year.values),loc='upper center',bbox_to_anchor=(1.05, 0.5),fancybox=True, shadow=True)
```

```python
#Distillate Fuel Oil
diesel_stk=eia_m('STEO.DFPSPUS.M','stock')[start:end]
diesel_prod=eia_m('STEO.DFROPUS.M','production')[start:end]
diesel_imp=eia_m('STEO.DFNIPUS.M','net imports')[start:end]
diesel_cons=eia_m('STEO.DFTCPUS.M','distillate fuel oil consumption')[start:end]

rcParams['figure.figsize'] = 16, 16
fig, axes = plt.subplots(nrows=4, ncols=1)
plot_seasonal_m(diesel_stk,axes[0],"mb")
plot_seasonal_m(diesel_prod,axes[1],"mb/d")
plot_seasonal_m(diesel_imp,axes[2],"mb/d")
plot_seasonal_m(diesel_cons,axes[3],"mb/d")
axes[0].set_title('U.S. Distillate Fuel Oil Projection, Monthly')
axes[0].legend(np.unique(diesel_prod.index.year.values),loc='upper center',bbox_to_anchor=(1.05, 0.5),fancybox=True, shadow=True)
```

```python
#Motor Gasoline
gasoline_stk=eia_m('STEO.MGTSPUS.M','stock')[start:end]
fgasoline_prod=eia_m('STEO.MGROPUS.M','finished motor gasoline production')[start:end]
fgasoline_imp=eia_m('STEO.MGNIPUS.M','finished motor gasoline net imports')[start:end]
bcgasoline_imp=eia_m('STEO.MBNIPUS.M','blending components net imports')[start:end]
gasoline_cons=eia_m('STEO.MGTCPUSX.M','total consumption')[start:end]

rcParams['figure.figsize'] = 16, 20
fig, axes = plt.subplots(nrows=5, ncols=1)
plot_seasonal_m(gasoline_stk,axes[0],"mb")
plot_seasonal_m(fgasoline_prod,axes[1],"mb/d")
plot_seasonal_m(fgasoline_imp,axes[2],"mb/d")
plot_seasonal_m(bcgasoline_imp,axes[3],"mb/d")
plot_seasonal_m(gasoline_cons,axes[4],"mb/d")
axes[0].set_title('U.S. Motor Gasoline Projection, Monthly')
axes[0].legend(np.unique(gasoline_stk.index.year.values),loc='upper center',bbox_to_anchor=(1.05, 0.5),fancybox=True, shadow=True)
```
